1. If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.


```python
n = 1000
sum = 0
for x in range(n):
    if x % 3 == 0 or x % 5 == 0:
        sum += x
    
print(sum)
```

3. The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?


```python
import math
def maxPrimeFactor(n):
   # number must be even
   while n % 2 == 0:
      max_Prime = 2
      n /= 1
   # number must be odd
   for i in range(3, int(math.sqrt(n)) + 1, 2):
      while n % i == 0:
         max_Prime = i
         n = n / i
   # prime number greator than two
   if n > 2:
      max_Prime = n
   return int(max_Prime)
# Driver code to test above function
n = 600851475143
print(maxPrimeFactor(n))
```

Find the largest palindrome made from the product of two 3-digit numbers.

8. A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
a^2 + b^2 = c^2

For example, 3^2 + 4^2 = 9 + 16 = 25 = 5^2.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.


```python
def find_pythagorean_triplet_product(sum_of_abc):
    for a in range(1, sum_of_abc):
        for b in range(a + 1, sum_of_abc):
            c = sum_of_abc - a - b
            if c > 0 and a**2 + b**2 == c**2:
                return a * b * c
    return None

print(find_pythagorean_triplet_product(1000))

```

9. The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.


```python
def sum_of_primes_below(n):
    nums = list(range(2, n))

    # Use Sieve of Eratosthenes to mark non-prime numbers
    for i in range(2, int(n ** 0.5) + 1):
        for j in range(i**2, n, i):
            nums[j - 2] = 0

    return sum(filter(lambda x: x != 0, nums))

print(sum_of_primes_below(2000000))
```

What is the value of the first triangle number to have over five hundred divisors?


```python
def count_divisors(n):
    """
    Returns the number of divisors of n.
    """
    count = 0
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            count += 2
    return count

def find_triangle_number_with_n_divisors(n):
    """
    Returns the first triangle number to have over n divisors.
    """
    divisors = 0
    i = 1
    triangle_number = 0
    while divisors <= n:
        triangle_number += i
        i += 1
        divisors = count_divisors(triangle_number)
    return triangle_number

# Find the first triangle number with over 500 divisors
n = 500
result = find_triangle_number_with_n_divisors(n)
print(result)
```
